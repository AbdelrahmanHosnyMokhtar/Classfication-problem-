```python
import pandas as pd 
import numpy as np
from IPython.display import display # Allows the use of display() for DataFrames

# Pretty display for notebooks
%matplotlib inline
#fun to return values to numeric and fill the NAN values 
def to_num(column):
    column = pd.to_numeric(column,errors = 'coerce').fillna(method='ffill')
    return column

training_data   = pd.read_csv("training.csv"  , sep=';')
validation_data = pd.read_csv("validation.csv", sep=';')
training_data['variable2'] =to_num(training_data['variable2'])
training_data['variable3'] =to_num(training_data['variable2'])
training_data['variable8'] =to_num(training_data['variable2'])
validation_data['variable2'] = to_num(validation_data['variable2'])
validation_data['variable3'] = to_num(validation_data['variable3'])
validation_data['variable8'] = to_num(validation_data['variable8'])
training_data = training_data.fillna(0)
validation_data = validation_data.fillna(0)
display(training_data.head(5))
```

```python
#training data prepration
from sklearn.preprocessing import MinMaxScaler
Label_t    = training_data['classLabel'] #the label column
features_t = training_data.drop('classLabel',axis = 1) #features columns
scaler   = MinMaxScaler() 
numrical = ['variable2', 'variable3', 'variable8', 'variable11', 'variable14', 'variable15', 'variable17', 'variable19']
features_minmax_transform_t = pd.DataFrame(data = features_t)
features_minmax_transform_t[numrical] = scaler.fit_transform(features_t[numrical])
final_features_t  = pd.get_dummies(features_minmax_transform_t)
final_features_t  = final_features_t.drop(['variable4_l','variable5_gg','variable6_r','variable7_o'],axis=1)
Label_t    = Label_t.replace({'no.':0,'yes.':1})
encoded = list(final_features_t.columns)
print("{} total features after one-hot encoding.".format(len(encoded)))
print(encoded)
#validation data prepration
Label_v    = validation_data['classLabel'] #the label column
features_v = validation_data.drop('classLabel',axis = 1) #features columns 
features_minmax_transform_v = pd.DataFrame(data = features_v)
features_minmax_transform_v[numrical] = scaler.fit_transform(features_v[numrical])
final_features_v  = pd.get_dummies(features_minmax_transform_v)
Label_v    = Label_v.replace({'no.':0,'yes.':1})
#The training and validation X and y
X_train = final_features_t
y_train = Label_t
X_valid = final_features_v
y_valid = Label_v
```

```python
#importing the algorithm and the accuracy function 
from sklearn.metrics import accuracy_score, fbeta_score
from sklearn.neighbors import KNeighborsClassifier

#train the model 
model = KNeighborsClassifier(n_neighbors=4)
model.fit(X_train, y_train)
predicted = model.predict(X_valid)
print('The accuracy score is ',accuracy_score(y_valid, predicted))
print('The F_beta_score is',fbeta_score(y_valid,predicted,1))
```
